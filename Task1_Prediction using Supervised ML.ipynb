## Task is to predict the percentage of a student based on the number of study hours.

### Import all the required libraries

```python
import pandas as pd # to read the data set 
import numpy as np # to deal with mathematical calculations
import matplotlib.pyplot as plt #In order to interpret visually
```

### Read the dataset:

```python
Data = pd.read_csv("http://bit.ly/w-data")
print("Data set has following dimensions:",Data.shape)
```

```python
Data.head()
```

### Information about type of dataset columns

```python
Data.info()
```

```python
Data.plot(x = 'Hours', y = 'Scores', style = 'o')
plt.title('Relation between hours studied and Scores scored')
plt.xlabel(" Hours spend by students (in Hours)")
plt.ylabel(" Scores scored by students (in percentages)")
plt.show()
```

**From the above plot, it is very clear that the score scored by student is linearly related to number of hours spend studying.** 

**NOTE: Here target column is scores scored by students (in Percentages), which is  continous variable.**

**Also, Scores is continous variable building a linear regression model is best choice.**

### Model building

Here, ols function(Ordinary Least Square function) is used to compute linear regression model. This ols function is generally used in context of data set that has variables which has direct proportional relation. 

```python
import statsmodels
import statsmodels.formula.api as sm
Model = sm.ols(formula = ' Scores~Hours',data = Data)

Fitted_Model = Model.fit()
Fitted_Model.summary()
```

####  1. R-squared value for this model is 0.953. This model has predicted 95% of data accurate.
####  2.Adjusted R squared value is 0.951. This is very close to R squared hence model is neither overfitted nor underfitted.
####  3.In terms of regression line given by Scores = beta_0  + beta_1 * Hours, beta_0 = 2.4837 and beta_1 = 9.7758

```python
Regression_Line = 2.4837 + 9.7758*Data.Hours
Data.plot(x = 'Hours', y = 'Scores', style = 'o')
plt.title('Relation between hours studied and Scores scored')
plt.xlabel(" Hours spend by students(in Hours)")
plt.ylabel(" Scores scored by students(in percentages)")
plt.plot(Data.Hours,Regression_Line)
plt.show()
```
